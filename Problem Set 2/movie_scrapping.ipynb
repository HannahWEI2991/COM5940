```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
from random import randint
from time import time
```

```python
header = ['电影名','链接','主演','上映时间']
data = []
```

```python
for i in range(0,10):
    quote_page = requests.get('https://maoyan.com/board/4?offset='+str(i * 10))
    sleep(randint(8,15))
    print("\n***** Page " + str(i+1) +" in action *****")
    soup = BeautifulSoup(quote_page.content,'html.parser')
    
    for article in soup.find_all("div", attrs={"class":"movie-item-info"}):
        page_no = str(i+1)
        title = article.a.text.encode('utf-8').strip()
        decoded_title = title.decode('utf-8')        
        print(decoded_title)
        url = article.a.get('href')
        movie_link = 'https://maoyan.com'+ url
        print(movie_link)
        all_p_tag = article.find_all('p')
        star = all_p_tag[1].text.encode('utf-8').strip()
        decoded_star = star.decode('utf-8')
        leadstar = decoded_star.strip('主演：')
        print(leadstar)
        time = all_p_tag[2].text
        date = time[5:14]
        print(date)
        data.append((decoded_title, movie_link, leadstar,date))
        
df = pd.DataFrame(data,
    columns = header
)
df.to_csv('MaoyanTop100movies.csv', sep='\t', encoding='utf-8')
```
